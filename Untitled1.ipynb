```python
def predict_disease(symptoms_list):
    """
    Input  : ['fever', 'headache', 'fatigue']
    Output : disease name + confidence %
    """
    # Load saved files
    model = pickle.load(open('disease_model.pkl', 'rb'))
    le = pickle.load(open('label_encoder.pkl', 'rb'))
    columns = pickle.load(open('symptom_columns.pkl', 'rb'))

    # Create input vector
    input_data = pd.DataFrame([{col: 0 for col in columns}])
    for symptom in symptoms_list:
        symptom = symptom.strip().lower().replace(' ', '_')
        if symptom in input_data.columns:
            input_data[symptom] = 1

    # Predict
    prediction = model.predict(input_data)[0]
    probabilities = model.predict_proba(input_data)[0]
    confidence = round(max(probabilities) * 100, 2)
    disease_name = le.inverse_transform([prediction])[0]

    return {
        "disease": disease_name,
        "confidence": f"{confidence}%"
    }

# Test it
result = predict_disease(['fever', 'headache', 'fatigue'])
print(result)
# Output: {'disease': 'Malaria', 'confidence': '94.5%'}
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
import pickle

# ── Load Dataset 1 ──────────────────────────────────────────
df1 = pd.read_csv('dataset1/dataset.csv')

# Clean symptom values (strip spaces)
symptom_cols = [c for c in df1.columns if 'Symptom' in c]
for col in symptom_cols:
    df1[col] = df1[col].str.strip()

# Get all unique symptoms
all_symptoms = pd.unique(df1[symptom_cols].values.ravel())
all_symptoms = sorted([s for s in all_symptoms if str(s) != 'nan'])
print(f"✅ Total unique symptoms found: {len(all_symptoms)}")

# Create binary columns for each symptom
for symptom in all_symptoms:
    df1[symptom] = df1[symptom_cols].apply(
        lambda row: 1 if symptom in row.values else 0, axis=1
    )

# Drop original symptom columns
df1_clean = df1.drop(columns=symptom_cols)
df1_clean = df1_clean.rename(columns={'Disease': 'disease'})

# ── Prepare Features & Target ────────────────────────────────
X1 = df1_clean.drop(columns=['disease'])
y1 = df1_clean['disease']

le1 = LabelEncoder()
y1_encoded = le1.fit_transform(y1)

# ── Train/Test Split ─────────────────────────────────────────
X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1_encoded, test_size=0.2, random_state=42, stratify=y1_encoded
)

# ── Train Random Forest ──────────────────────────────────────
model1 = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1
)
model1.fit(X1_train, y1_train)

# ── Evaluate ─────────────────────────────────────────────────
y1_pred = model1.predict(X1_test)
acc1 = accuracy_score(y1_test, y1_pred)
print(f"\n🎯 Model 1 Accuracy (Symptom): {acc1 * 100:.2f}%")
print("\n📊 Classification Report:")
print(classification_report(y1_test, y1_pred, target_names=le1.classes_))

# ── Save Model ───────────────────────────────────────────────
pickle.dump(model1, open('symptom_model.pkl', 'wb'))
pickle.dump(le1, open('symptom_label_encoder.pkl', 'wb'))
pickle.dump(list(X1.columns), open('symptom_columns.pkl', 'wb'))
print("\n✅ Model 1 saved successfully!")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
import pickle

# ── Load Dataset 2 ──────────────────────────────────────────
df2 = pd.read_csv('dataset2/Blood_samples_dataset_balanced_2(f).csv')

# Fix truncated disease names
disease_map = {
    'Thalasse': 'Thalassemia',
    'Thromboc': 'Thrombocytopenia',
    'Healthy' : 'Healthy',
    'Anemia'  : 'Anemia',
    'Diabetes': 'Diabetes'
}
df2['Disease'] = df2['Disease'].str.strip().map(disease_map)
print("✅ Disease names fixed:")
print(df2['Disease'].value_counts())

# ── Prepare Features & Target ────────────────────────────────
X2 = df2.drop(columns=['Disease'])
y2 = df2['Disease']

le2 = LabelEncoder()
y2_encoded = le2.fit_transform(y2)

# ── Train/Test Split ─────────────────────────────────────────
X2_train, X2_test, y2_train, y2_test = train_test_split(
    X2, y2_encoded, test_size=0.2, random_state=42, stratify=y2_encoded
)

# ── Train Random Forest ──────────────────────────────────────
model2 = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42,
    n_jobs=-1
)
model2.fit(X2_train, y2_train)

# ── Evaluate ─────────────────────────────────────────────────
y2_pred = model2.predict(X2_test)
acc2 = accuracy_score(y2_test, y2_pred)
print(f"\n🎯 Model 2 Accuracy (Blood Report): {acc2 * 100:.2f}%")
print("\n📊 Classification Report:")
print(classification_report(y2_test, y2_pred, target_names=le2.classes_))

# ── Save Model ───────────────────────────────────────────────
pickle.dump(model2, open('blood_model.pkl', 'wb'))
pickle.dump(le2, open('blood_label_encoder.pkl', 'wb'))
pickle.dump(list(X2.columns), open('blood_columns.pkl', 'wb'))
print("\n✅ Blood model saved successfully!")
```

```python
# ── Real World Ranges for Normalization ──────────────────────
BLOOD_RANGES = {
    'Glucose'                                  : (70,  200),
    'Cholesterol'                              : (100, 300),
    'Hemoglobin'                               : (7,   20),
    'Platelets'                                : (50,  450),
    'White Blood Cells'                        : (2,   20),
    'Red Blood Cells'                          : (2,   7),
    'Hematocrit'                               : (20,  60),
    'Mean Corpuscular Volume'                  : (60,  120),
    'Mean Corpuscular Hemoglobin'              : (10,  40),
    'Mean Corpuscular Hemoglobin Concentration': (25,  40),
    'Insulin'                                  : (2,   300),
    'BMI'                                      : (10,  50),
    'Systolic Blood Pressure'                  : (80,  200),
    'Diastolic Blood Pressure'                 : (50,  130),
    'Triglycerides'                            : (50,  500),
    'HbA1c'                                    : (4,   15),
    'LDL Cholesterol'                          : (50,  250),
    'HDL Cholesterol'                          : (20,  100),
    'ALT'                                      : (5,   100),
    'AST'                                      : (5,   100),
    'Heart Rate'                               : (40,  150),
    'Creatinine'                               : (0.4, 10),
    'Troponin'                                 : (0,   3),
    'C-reactive Protein'                       : (0,   200)
}

def normalize_blood_values(raw_values):
    """
    Converts real-world blood values to 0-1 scale
    Input  : {'Glucose': 126, 'Hemoglobin': 9.5, ...}
    Output : {'Glucose': 0.42, 'Hemoglobin': 0.19, ...}
    """
    normalized = {}
    for param, value in raw_values.items():
        if param in BLOOD_RANGES:
            min_val, max_val = BLOOD_RANGES[param]
            # Clip value within range then normalize
            clipped = max(min_val, min(max_val, value))
            normalized[param] = round(
                (clipped - min_val) / (max_val - min_val), 6
            )
        else:
            print(f"⚠️ Unknown parameter: {param}")
    return normalized


def predict_by_blood_raw(raw_blood_values):
    """
    Takes REAL blood report values (not normalized)
    Input  : {'Glucose': 126, 'HbA1c': 7.5, 'Hemoglobin': 9.5 ...}
    Output : {'disease': 'Diabetes', 'confidence': '92.5%', 'top3': [...]}
    """
    model   = pickle.load(open('blood_model.pkl', 'rb'))
    le      = pickle.load(open('blood_label_encoder.pkl', 'rb'))
    columns = pickle.load(open('blood_columns.pkl', 'rb'))

    # Step 1 — Normalize real values to 0-1
    normalized = normalize_blood_values(raw_blood_values)

    # Step 2 — Build full input vector (fill missing with 0.5 = midrange)
    input_data = pd.DataFrame([{
        col: normalized.get(col, 0.5) for col in columns
    }])

    # Step 3 — Predict
    pred       = model.predict(input_data)[0]
    proba      = model.predict_proba(input_data)[0]
    confidence = round(max(proba) * 100, 2)
    disease    = le.inverse_transform([pred])[0]

    # Step 4 — Top 3 predictions
    top3_idx = proba.argsort()[-3:][::-1]
    top3     = [(le.inverse_transform([i])[0],
                 round(proba[i] * 100, 2)) for i in top3_idx]

    return {
        "disease"         : disease,
        "confidence"      : f"{confidence}%",
        "top3"            : top3,
        "normalized_input": normalized  # useful for debugging
    }
```
